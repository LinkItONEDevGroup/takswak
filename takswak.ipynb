```python

def cot_gen(cotd,cot_type="marker"): # cotd: untangle 的結果
    """
    Generates COT Event by cotd. 使用方法是找個範例的 xml 轉出 cotd, 修改參數後來使用
    FIXME: stale 時間寫死
    """
    root = ET.Element("event")
    root.set("version", cotd.event['version'])
    root.set("uid", cotd.event['uid'])
    root.set("type", cotd.event['type'])
    if 'how' in cotd.event._attributes:
        root.set("how", cotd.event['how'])
    root.set("time", pytak.cot_time())
    root.set("start", pytak.cot_time())
    root.set("stale", pytak.cot_time(60*10))
    if 1:
        if hasattr(cotd.event,'point'):
            point = ET.SubElement(root, 'point')
            point.set("lat",cotd.event.point['lat'])
            point.set("lon",cotd.event.point['lon'])
            point.set("ce",cotd.event.point['ce'])
            point.set("hae",cotd.event.point['hae'])
            point.set("le",cotd.event.point['le'])
        if 1:
            
            if hasattr(cotd.event,'detail'):
                detail = ET.SubElement(root, 'detail')
                if hasattr(cotd.event.detail,'contact') and 1:
                    contact = ET.SubElement(detail, 'contact')
                    contact.set("callsign",cotd.event.detail.contact['callsign'])
                    if 'endpoint' in cotd.event.detail.contact._attributes:
                        contact.set("endpoint",cotd.event.detail.contact['endpoint'])
                if hasattr(cotd.event.detail,'uid') and 1:
                    uid = ET.SubElement(detail, 'uid')
                    uid.set("Droid",cotd.event.detail.uid['Droid'])
                if hasattr(cotd.event.detail,'precisionlocation') and 1:
                    precisionlocation = ET.SubElement(detail, 'precisionlocation')
                    precisionlocation.set("altsrc",cotd.event.detail.precisionlocation['altsrc'])
                    if 'geopointsrc' in cotd.event.detail.precisionlocation._attributes:
                        precisionlocation.set("geopointsrc",cotd.event.detail.precisionlocation['geopointsrc'])
                if hasattr(cotd.event.detail,'__group') and 1:
                    __group = ET.SubElement(detail, '__group')
                    __group.set("role",cotd.event.detail.__group['role'])
                    __group.set("name",cotd.event.detail.__group['name'])
                if hasattr(cotd.event.detail,'status') and 1:
                    status = ET.SubElement(detail, 'status')
                    if 'readiness' in cotd.event.detail.status._attributes:
                        status.set("readiness",cotd.event.detail.status['readiness'])
                    if 'battery' in cotd.event.detail.status._attributes:
                        status.set("battery",cotd.event.detail.status['battery'])
                if hasattr(cotd.event.detail,'track') and 1:
                    track = ET.SubElement(detail, 'track')
                    if 'course' in cotd.event.detail.track._attributes:
                        track.set("course",cotd.event.detail.track['course'])
                        track.set("speed",cotd.event.detail.track['speed'])
                

                remarks = ET.SubElement(detail, 'remarks')
                if hasattr(cotd.event.detail,'remarks') and 1:
                    remarks.text=cotd.event.detail.remarks.cdata
                    #print("remarks set=%s" %(cotd.event.detail.remarks))
                if hasattr(cotd.event.detail,'usericon'):
                    usericon = ET.SubElement(detail, 'usericon')
                    usericon.set("iconsetpath",cotd.event.detail.usericon['iconsetpath'])
                if hasattr(cotd.event.detail,'link') and 1:
                    link = ET.SubElement(detail, 'link')
                    link.set("uid",cotd.event.detail.link['uid'])
                    link.set("production_time",pytak.cot_time())
                    link.set("type",cotd.event.detail.link['type'])
                    link.set("parent_callsign",cotd.event.detail.link['parent_callsign'])
                    link.set("relation",cotd.event.detail.link['relation'])
    return ET.tostring(root)

```

```python
""" hardware demo
密碼 對象 時間 命令 參數
pwd microbit 20230301153000 fire x y
"""

import serial
from threading import Timer
from datetime import datetime

ser=None #serial port
def readall(ser,verbose=True):
    lines=[]
    while True:
        line = ser.readline()
        if line==b"":
            break
        else:
            if verbose: print(line) 
            lines.append(line)
    return lines
def run_hwcmd(hw_cmd,verbose=True):
    ok=0
    global ser
    if not ser:
        ser = serial.Serial('/dev/tty.usbmodem2102',baudrate=115200,timeout=0.1)
    readall(ser,verbose)
    
    ser.write(hw_cmd.encode())
    ser.write([0xa,0xd])
    time.sleep(0.1)
    lines = readall(ser,verbose)
    for line in lines:
        linestr= str(line)
        if linestr.find("OK:"+hw_cmd)>=0:
            ok=1
    return ok


_hw_cmd=""
def timer_fun():
    global _hw_cmd
    run_hwcmd(_hw_cmd,True)
    


def cal_timediff(end_time):
    #end_time = "20230302105000"

    # convert time string to datetime
    t1 = datetime.now()
    #print('Start time:', t1.time())

    t2 = datetime.strptime(end_time, "%Y%m%d%H%M%S")
    #print('End time:', t2.time())

    # get difference
    delta = t2 - t1

    # time difference in seconds
    #print(f"Time difference is {delta.total_seconds()} seconds")
    return delta.total_seconds()


def run_takcmd(tak_cmd):
    cols = tak_cmd.split()
    if len(cols)>=1:
        if cols[0]=="pwd":
            if len(cols)>=2:
                if cols[1]=="microbit":
                    if len(cols)>=3:
                        cmd_time = cols[2]
                        if len(cols)>=4:
                            hw_cmd = " ".join(cols[3:])
                            print("time=%s,hw_cmd=%s" %(cmd_time,hw_cmd))
                            if cmd_time=="now":
                                ok=run_hwcmd(hw_cmd,True)
                            else:
                                #cal second
                                time_period = cal_timediff(cmd_time)
                                if time_period>=0:
                                    #time_period=3
                                    global _hw_cmd
                                    _hw_cmd = hw_cmd
                                    t = Timer(time_period, timer_fun)
                                    t.start()
                                else:
                                    return 0
                            return 1
    return 0
                            
#print(run_takcmd("pwd microbit 20230302104900 fire 121 24"))
                    
```

```python
import xml.etree.ElementTree as ET
def tak_pong():
    """
    Generates a simple takPong COT Event.
    """
    root = ET.Element("event")
    root.set("version", "2.0")
    root.set("type", "t-x-d-d")
    root.set("uid", "takPong")
    root.set("how", "m-g")
    root.set("time", pytak.cot_time())
    root.set("start", pytak.cot_time())
    root.set("stale", pytak.cot_time(3600))
    return ET.tostring(root)

print(tak_pong())
```

# 匯出 icon

```python
# atak 的 icon 在 sqlite 的 DB 檔案中，用這些 function 可以取得某個 icon 的 png, 並存出檔案
import sqlite3

def writeTofile(data, filename):
    # Convert binary data to proper format and write it on Hard Disk
    with open(filename, 'wb') as file:
        file.write(data)
    print("Stored blob data into: ", filename, "\n")

def readBlobData(id): # id: iconsets.sqlite icon 表格中的 id 欄位值
    try:
        sqliteConnection = sqlite3.connect('include/iconsets.sqlite')
        cursor = sqliteConnection.cursor()
        print("Connected to SQLite")

        sql_fetch_blob_query = """SELECT id,filename,bitmap from icons where id = ?"""
        cursor.execute(sql_fetch_blob_query, (id,))
        record = cursor.fetchall()
        for row in record:
            print("Id = ", row[0], "Name = ", row[1])
            name = row[1]
            photo = row[2]

            print("Storing icon image  on disk \n")
            photoPath = "output/" + name
            
            writeTofile(photo, photoPath)
            

        cursor.close()

    except sqlite3.Error as error:
        print("Failed to read blob data from sqlite table", error)
    finally:
        if sqliteConnection:
            sqliteConnection.close()
            print("sqlite connection is closed")

# 建議先使用 DB Browser for SQLite 查詢想要的代號
readBlobData(3209)
```

# push-cursor-on-target
github: https://github.com/sofwerx/push-cursor-on-target/blob/master/CoT.py

將專案簡化成這個
這裡有代號對應，方便查閱

```python
import datetime as dt
import uuid
import xml.etree.ElementTree as ET
import socket
import logging
import os
import time

#logger = logging.getLogger("django")

ID = {
    "pending": "p",
    "unknown": "u",
    "assumed-friend": "a",
    "friend": "f",
    "neutral": "n",
    "suspect": "s",
    "hostile": "h",
    "joker": "j",
    "faker": "f",
    "none": "o",
    "other": "x"
}
DIM = {
    "space": "P",
    "air": "A",
    "land-unit": "G",
    "land-equipment": "G",
    "land-installation": "G",
    "sea-surface": "S",
    "sea-subsurface": "U",
    "subsurface": "U",
    "other": "X"
}

DATETIME_FMT = "%Y-%m-%dT%H:%M:%SZ"

class CursorOnTarget:

    def atoms(__self, unit):
        timer = dt.datetime
        now = timer.utcnow()
        zulu = now.strftime(DATETIME_FMT)
        stale_part = now.minute + 1
        if stale_part > 59:
            stale_part = stale_part - 60
        stale_now = now.replace(minute=stale_part)
        stale = stale_now.strftime(DATETIME_FMT)

        unit_id = ID[unit["identity"]] or ID["none"]
    
        cot_type = "a-" + unit_id + "-" + DIM[unit["dimension"]]

        if "type" in unit:
          cot_type = cot_type + "-" + unit["type"]

        if "uid" in unit:
          cot_id = unit["uid"]
        else:
          cot_id = uuid.uuid4().get_hex()

        evt_attr = {
            "version": "2.0",
            "uid": cot_id,
            "time": zulu,
            "start": zulu,
            "stale": stale,
            "type": cot_type
        }

        pt_attr = {
            "lat": str(unit["lat"]),
            "lon":  str(unit["lon"]),
            "hae": "0",   #unit["hae"],
            "ce": "10",    #unit["ce"],
            "le": "10"     #unit["le"]1
        }
    
        cot = ET.Element('event', attrib=evt_attr)
        ET.SubElement(cot, 'detail')
        ET.SubElement(cot,'point', attrib=pt_attr)
    
        #cot_xml = '<?xml version="1.0" standalone="yes"?>' + ET.tostring(cot)
        #return cot_xml
        return ET.tostring(cot)

    def pushUDP(__self, ip_address, port, cot_xml):
        sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        sent = sock.sendto(cot_xml, (ip_address, port))
        return sent

    def pushTCP(__self, ip_address, port, cot_xml):
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        conn = sock.connect((ip_address, port))
        return sock.send(cot_xml)
    
    
ATAK_IP = os.getenv('ATAK_IP', '192.168.1.160')
ATAK_PORT = int(os.getenv('ATAK_PORT', '4242'))
ATAK_PROTO = os.getenv('ATAK_PROTO', 'UDP')

params = {  # SWX parking lot
    "lat": 27.957261,
    "lon": -82.436587,
    "uid": "Nerd Herd",
    "identity": "hostile",
    "dimension": "land-unit",
    "entity": "military",
    "type": "U-C"
#    "type": "U-C-R-H"
}

for i in range(0, 10):
    params["lat"] = params["lat"] + i/10000.0
    params["lon"] = params["lon"] + i/10000.0
    print("\nParams:\n" + str(params))
    cot = CursorOnTarget()
    cot_xml = cot.atoms(params)


    print("XML message:")
    print(cot_xml)
    
    if 0:
        print("\nPushing to ATAK...")
        if ATAK_PROTO == "TCP":
            sent = cot.pushTCP(ATAK_IP, ATAK_PORT, cot_xml)
        else:
            sent = cot.pushUDP(ATAK_IP, ATAK_PORT, cot_xml)
        print(str(sent) + " bytes sent to " + ATAK_IP + " on port " + str(ATAK_PORT))
    time.sleep(10)
    
```
